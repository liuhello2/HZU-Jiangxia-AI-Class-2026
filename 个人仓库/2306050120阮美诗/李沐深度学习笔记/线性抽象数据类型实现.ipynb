栈：Stack() push(x) pop() peek() isEmpty() size()

```python
class Stack:
    def __init__(self):
        self.items=[]
    
    def isEmpty(self):
        return self.items==[]
    
    def push(self,item):
        self.items.append(item)
    
    def pop(self):
        return self.items.pop()
    
    def peek(self):
        return self.items[len(self.items)-1]
    
    def size(self):
        return len(self.items)
```

```python
S=Stack()
S.size()
```

队列：Queue() enqueue(item) dequeue() isEmpty() size()

```python
class Queue:
    def __init__(self):
        self.queue=[]
    
    def enqueue(self,item):
        self.queue.insert(0,item)
    
    def dequeue(self):
        return self.queue.pop()
    
    def isEmpty(self):
        return self.queue==[]
    
    def size(self):
        return len(self.queue)
```
