```python
##保证该定积分在[a,b]上每一点均有定义才能使用！！！##

from math import e,pi,sin,cos,tan,asin,acos,atan,log

def y(x,expression):
    return eval(expression)

expression=input("输入函数表达式（以x为自变量）：")
a=eval(input("积分上限："))
b=eval(input("积分下限："))
n=int(input("精度（分割数，输入正整数）："))
I=0
for i in range(1,n+1):
    x=((a-b)/n) *i+b
    xi=((a-b)/n) *(i-1)+b
    I=I+(y(x,expression)+y(xi,expression))*(a-b)/(2*n)
print("该积分的近似值为：",I)
```
